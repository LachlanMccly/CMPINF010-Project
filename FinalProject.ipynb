**Parks** - Lachlan

Park data is shown below. Imported the data from my csv file and used a loop to isolate the park-to-neighborhood ratio in the code.

```python
import pandas as pd
data = pd.read_csv("cityParks.csv",index_col="neighborhood")

counts = dict()

for index in data.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index]+=1



park_series = data.groupby("neighborhood").count().loc[:,"_id"].sort_values(ascending=False)

#Display the first 20
park_series[:20]
```

```python
park_series[:40].plot.bar()
```

# Conclusion
    East Liberty was the winner for city parks with 12 total parks. Each time a neighbordhood was called they got 1 point.

In this data set, there were 125 total playgrounds in these neighborhoods of Pittsburgh. 
- First, I found the total number of playgrounds
- Second, I found how many neighborhood had below 1 playground so I can rule them out
- Lastly, I found the highest playground count in a neighborhood

```python
import pandas as pd
```

```python
## Total number of playgrounds
data = pd.read_csv("CityPlaygrounds.csv")

total_playgrounds = data.shape[0]

print(f"Total number of playgrounds: {total_playgrounds}")
```

```python
## Display the neighborhoods, highest to lowest, with their respective amount of playgrounds
data = pd.read_csv("CityPlaygrounds.csv", index_col="neighborhood")

counts = dict()
for index in data.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

playground_series = data.groupby("neighborhood").count().loc[:,"_id"].sort_values(ascending=False)

# Display the first 20 to avoid clutter
playground_series[:20]
```

```python
## Display the number of neighborhoods with 1 or less playgrounds
data = pd.read_csv("CityPlaygrounds.csv")

playground_counts = data["neighborhood"].value_counts()

low_playground_neighborhoods = playground_counts[playground_counts <= 1]

num_low_playground_neighborhoods = len(low_playground_neighborhoods)

print("The number of neighbrhoods with 1 or less playgrounds:" + str(num_low_playground_neighborhoods))
```

```python
playground_series[:40].plot.bar()
```

**Combining the Data**
-
Finally, we combined the data together to find the neighborhood that had the most parks, pools, and playgrounds

```python
import pandas as pd 
import matplotlib.pyplot as plt

dataPools = pd.read_csv("Pools.csv",index_col = "neighborhood")
dataParks = pd.read_csv("cityParks.csv",index_col="neighborhood")
dataPlaygrounds = pd.read_csv("CityPlaygrounds.csv", index_col="neighborhood")

counts = dict()


for index in dataPlaygrounds.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

for index in dataPools.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

for index in dataParks.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

sorted_counts = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))


print(sorted_counts)

```

- This simple dictionary with the data sets counts which neighborhood has the most total of Pools, Parks, and Playgrounds 
- sorted_counts is the dictionary with the amounts of each neighborhoods mentioned from greatest to least so that the first index of the dictionary is the neighborhood with the most amount also known as the best neighborhood in Pittsburgh and in this case it was East Liberty

```python

sorted_counts_series = pd.Series(sorted_counts)

sorted_counts_series.plot.bar(figsize=(30, 4), color='skyblue')

plt.title("Neighborhoods Ranked Best to Worst")
plt.xlabel("Neighborhood")  
plt.ylabel("Number of parks, playgrounds, and pools")      


plt.show()
```

**Consensus**
-
This bar graph shows each neighborhood in Pittsburgh and the amount of total pools, parks, and playgrounds they have. It shows a clear **winner in East Liberty** and a clear **loser with Shadyside**. Having a limited amount of pools, parks, and playgrounds means no cheap fun that is usually in walking distance. Overall, we believe that in East Liberty, you can have the most **family fun outside**.
