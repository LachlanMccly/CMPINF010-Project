Our goal was to find the best neighborhood in Pittsburgh by seeing which neighborhood had the most pools, parks, and playgrounds. We used this metric as kids need these 3 community areas to find friends, have fun, as well as building a stronger community

```python
import pandas as pd 
import matplotlib.pyplot as plt

dataPools = pd.read_csv("Pools.csv",index_col = "neighborhood")
dataParks = pd.read_csv("cityPark.csv",index_col="neighborhood")
dataPlaygrounds = pd.read_csv("cityPlay.csv", index_col="neighborhood")

counts = dict()


for index in dataPlaygrounds.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

for index in dataPools.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

for index in dataParks.index:
    if index not in counts:
        counts[index] = 1
    else:
        counts[index] += 1

sorted_counts = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))


print(sorted_counts)

```

This simple dictionary with the data sets counts which neighborhood has the most total of Pools, Parks, and Playgrounds 
sorted_counts is the dictionary with the amounts of each neighborhoods mentioned from greatest to least so that the first index of the dictionary is the neighborhood with the most amount also known as the best neighborhood in Pittsburgh and in this case it was East Liberty

```python

sorted_counts_series = pd.Series(sorted_counts)

sorted_counts_series.plot.bar(figsize=(30, 4), color='skyblue')

plt.show()
```

This bar graph shows each neighborhood in Pittsburgh and the amount of total pools, parks, and playgrounds they have. It shows a clear winner in East Liberity and a clear loser with Shadyside. Having a limited amount of pools, parks, and playgrounds means no cheap fun that is usually in walking distance.
